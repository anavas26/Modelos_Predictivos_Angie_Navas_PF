```python

#
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# lectura de archivo
df = pd.read_csv("/content/world food production.csv")

#
df.head()

#
df_panama = df[df['Entity'] == 'Americas (FAO)']

#
df_panama.head()

#
df.columns

#
products = [
       'Rice  Production ( tonnes)', 'Yams  Production (tonnes)',
        'Tomatoes Production (tonnes)', 'Sugar cane Production (tonnes)',
       'Potatoes  Production (tonnes)', 'Coffee, green Production ( tonnes)',
       'Bananas  Production ( tonnes)', 'Meat, chicken  Production (tonnes)']

#
for product in products:
    plt.figure(figsize=(14, 10))
    plt.plot(df_panama['Year'], df_panama[product], linestyle="-", color='b')
    plt.title(product)
    plt.xlabel("Years")
    plt.ylabel("Tonnes")
    plt.grid(True)
    plt.show()

#
df_panama.columns

#
from statsmodels.tsa.seasonal import seasonal_decompose

df_panama['Year'] = pd.to_datetime(df_panama['Year'], format = '%Y')
df_panama.set_index('Year', inplace=True)
decomposition = seasonal_decompose(df_panama['Rice  Production ( tonnes)'],
                                   model='additive',
                                   period = 1)
decomposition.plot()
plt.show()

#
from statsmodels.tsa.stattools import adfuller

result = adfuller(df_panama['Rice  Production ( tonnes)'])
print('ADF Statistics: ', result[0])
print('p-value: ', result[1])

#
from statsmodels.tsa.holtwinters import ExponentialSmoothing

model = ExponentialSmoothing(df_panama['Rice  Production ( tonnes)'], trend='mul', seasonal='mul', seasonal_periods=10)
fitted_model = model.fit()

#
df_panama['Forecast'] = fitted_model.fittedvalues
df_panama[['Rice  Production ( tonnes)', 'Forecast']].plot(figsize=(12,8))

#
future = pd.date_range(start=df_panama.index[-1], periods=11, freq='Y')[1:]
forecast = fitted_model.forecast(steps=10)

#
forecast_df = pd.DataFrame({'Year': future, 'Forecast':forecast.values})
forecast_df.set_index('Year', inplace=True)

plt.figure(figsize=(12,6))
plt.plot(df_panama.index, df_panama['Rice  Production ( tonnes)'], label='Produccion Actual')
plt.plot(forecast_df.index, forecast_df['Forecast'], label='Next 10 years')

#
df_panama.head()
```
